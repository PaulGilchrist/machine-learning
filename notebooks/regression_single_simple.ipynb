# Linear and Polinormial Regression

Predict the best Y for any given X (linear regression) using existing package

```python
# %pip install matplotlib
# %pip install numpy
# %pip install scipy
# %pip install sklearn

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats  # built in linear regression capability
from sklearn.metrics import r2_score

# We are defaulting to CPU, but you may want to compare CPU vs GPU performance
# device = "mps" if torch.backends.mps.is_available() else "cuda" if torch.cuda.is_available() else "cpu"
device = "cpu" # Use cpu since this dataset is too small to take advantage of the GPU
```

```python
# Generate mock data
x = [4.7,2.4,7.5,7.1,4.3,7.816,8.9,5.2,8.59,2.1,8,10,4.5,6,4]
y = [2.6,1.6,3.09,2.4,2.4,3.357,2.6,1.96,3.53,1.76,3.2,3.5,1.6,2.5,2.2]

# Display Scattter Plot
plt.figure(figsize=(12,8))
plt.scatter(x, y, label='Original data', s=250, c='g')
plt.legend()
plt.title("Scatter Plot Training Data")
plt.show()
```

```python
slope, intercept, relationship, p, std_err = stats.linregress(x, y)
# A low value of `relationship` indicates a bad fit for linear regression and may be better to try polynomial regression
print(relationship)

def linearRegression(x):
  return slope * x + intercept

model_linear = list(map(linearRegression, x))

# Display Scattter Plot with Linear Regression
plt.figure(figsize=(12,8))
plt.scatter(x, y, label='Original data', s=250, c='g')
plt.plot(x, model_linear, 'r', label='Fitted line')
plt.legend()
plt.title("Linear Regression")
plt.show()
```

```python
model_poly = np.poly1d(np.polyfit(x, y, 3))
relationship = r2_score(y, model_poly(x))
# A low value of `relationship` indicates a bad fit for polynomial regression
print(relationship)
line = np.linspace(1, max(x), 100)

# Display Scattter Plot with Polynomial Regression
plt.figure(figsize=(12,8))
plt.scatter(x, y, label='Original data', s=250, c='g')
plt.plot(line, model_poly(line), 'r', label='Fitted line')
plt.legend()
plt.title("Polynomial Regression")
plt.show()
```
